Mutable and Immutable Data Types
But lists and strings are different in an important way. A list value is a mutable data type: It can have values added, removed, or changed. However, a string is immutable: It cannot be changed. Trying to reassign a single character in a string results in a TypeError error, as you can see by entering the following into the interactive shell

```python
name = 'Zophie a c  at'
name[7] = 'the'
```

The proper way to “mutate” a string is to use slicing and concatenation to build a new string by copying from parts of the old string. Enter the following into the interactive shell:


```python
name = 'Zophie a cat'
newName = name[0:7] + 'the' + name[8:12]
print(name)
print(newName)
```

```python
def f(iman=[]):
    
    iman.append(1)
    print("In function -",id(iman), "- ", iman)
 
    return iman
   
print(id(iman), "- ", iman)
print()

f() # 1st calling
print("out function -",id(iman), "- ", iman)
print()


f() # 2nd calling
print("out function -",id(iman), "- ", iman)
print()

f(iman=[2]) # 3rd calling
print("out function -",id(iman), "- ", iman)
print()


f() # 4th calling
print("out function -",id(iman), "- ", iman)
```

```python
The reason for the error -

When a.show("bmw x4", "red") is executed,

Python convert it into car.show(a, "bmw x4", "red") # class_name.method_name(instance, argument1, argument2, ...)

Now, the call 'car.show(a, "bmw x4", "red")' has three arguments such as,
1. a - instance of the class
2. "bmw x4" - model 
3. "red" - color


But, in defining the show() method in class car, we just gave only two argument(model and, color) - "def show(model, color):"

To resolve this error, we add one more argument in the method definition to represent the instance. Like -"def show(self, model, color):" 

Now, the call 'car.show(a, "d", "e")' has three arguments and, also the definition 'def show(self, model, color):' has three arguments

Number of arguments matched and, works fine
```

###### Inheritance

```python
class parent(object):
    # Constructor 
    def __init__(self, name):
        self.name = name 
    
    def getName(self): 
        return self.name 

    def boolean(self): 
        return False
   
   
# Inherited or Sub class
class child(parent): 
    # Here we return true 
    def boolean(self): 
        return True
    
    emp = parent("Geek1")
    print(emp.getName(), emp.boolean()) 

emp = child("Geek2")
print(emp.getName(), emp.boolean())
```

```python
class parent( object ):     
    
    # __init__ is known as the constructor          
    def __init__(self, name, address):    
        self.name = name 
        self.address = address 
    def display(self):
        print(self.name) 
        print(self.address)
        print("parent class")
        
class child( parent ):            
        def __init__(self, name, address, age, height): 
                self.name = name 
                self.address = address
                self.age = age 
                self.height = height 
  
  
                  
# creation of an object variable or an instance 
a = child('Ram', "India", 3, 50)    
  
a.display()  
```
